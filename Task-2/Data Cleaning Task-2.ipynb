# Import Libraries & Load the Dataset

```python
# Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("AB_NYC_2019.csv")
df
```

# Standardize Column Names

```python
# Standardize column names: lowercase, replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Check updated columns
print("Updated column names:\n", df.columns.tolist())
```

# Handling Missing Values

```python
# Count total missing values
missing = df.isnull().sum()
missing
```

```python
# or in other way..
# Count total missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print("Missing Values in Each Column:\n", missing)
```

```python
# Calculate percentage of missing
missing_percent = (missing / len(df)) * 100
print("\nPercentage of Missing Values:\n", missing_percent)
```

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
```

# Filling missing values where appropriate

```python
# For 'name' and 'host_name', fill with 'Unknown'

df['name'].fillna('Unknown', inplace=True)

df['host_name'].fillna('Unknown', inplace=True)
```

```python
# For 'reviews_per_month', fill missing with 0

df['reviews_per_month'].fillna(0, inplace=True)

# Drop any remaining missing rows if minor
df.dropna(inplace=True)
```

```python
print("Missing Values After Cleaning:")

print(df.isnull().sum())

# now we don't have any missing values in the dataset... we fill the null values
```

# Handling Duplicates Values

```python
# Check for duplicates

duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Verify removal
print("Duplicates removed. Remaining rows:", df.shape[0])
```

# Data Type Correction

```python
print(df.dtypes)
```

```python
# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
print("Data types after conversion:")
print(df.dtypes)
```

# Text & Categorical Standardization

```python
# Standardize categorical text fields

df['neighbourhood_group'] = df['neighbourhood_group'].str.title()

df['neighbourhood'] = df['neighbourhood'].str.title()

df['room_type'] = df['room_type'].str.title()

# Check unique categories

print("Neighbourhood groups:", df['neighbourhood_group'].unique())

print("Room types:", df['room_type'].unique())
```

# Outlier Detection ( IQR Method )

```python
num_cols = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)   # 25th percentile
    Q3 = df[col].quantile(0.75)   # 75th percentile
    IQR = Q3 - Q1                 # Spread of middle 50%
    lower = Q1 - 1.5 * IQR        # Minimum limit
    upper = Q3 + 1.5 * IQR        # Maximum limit
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```
