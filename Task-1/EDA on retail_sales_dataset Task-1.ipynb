# 2. Descriptive Statistics

```python
# BASIC STATISTICS

# 1.Calculate the Mean (Average) of numeric columns

mean_values = df[['Age','Quantity','Price per Unit','Total Amount']].mean()
print("Mean Values:\n",mean_values)
```

```python
# 2.Calculate the Median of numeric columns

median_values = df[['Age','Quantity','Price per Unit','Total Amount']].median()
print("Median values:\n",median_values)
```

```python
# 3. Calculate the Mode (repeated values)

mode_gender = df['Gender'].mode()[0]
print("Most common gender:\n", mode_gender ,"\n")

mode_category = df['Product Category'].mode()[0]
print("Most common product category:\n", mode_category, "\n")

mode_quantity = df['Quantity'].mode()[0]
print("Most common quantity purchased:\n", mode_quantity)
```

```python
# 4. Standard Deviation of numeric columns

std_dev = df[['Age','Quantity','Price per Unit','Total Amount']].std()
print("Standard deviation:\n", std_dev)
```

# 3. Time Series Analysis

```python
import matplotlib.pyplot as plt

# Monthly sales trend
monthly_sales = df.resample('ME', on='Date')['Total Amount'].sum()

# Plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
monthly_sales.plot(marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales (â‚¹)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Monthly Sales Trend Analysis

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Resample data to monthly frequency and calculate total sales
monthly_sales = df.resample('M', on='Date')['Total Amount'].sum()

# Calculate a 3-month moving average to smooth the trend
monthly_sales_MA = monthly_sales.rolling(window=3).mean()

# Plot the monthly sales trend and moving average
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, 
         marker='o', color='teal', label='Monthly Sales')
plt.plot(monthly_sales_MA.index, monthly_sales_MA.values, 
         color='red', linewidth=2, label='3-Month Moving Average')

# Title, labels, and formatting
plt.title('Monthly Sales Trend (with 3-Month Moving Average)', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales (â‚¹)', fontsize=12)

# Format x-axis to show readable month names
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Grid and legend for better readability
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()

# Display key summary insights

print("Sales Summary:")
print(f"Highest Monthly Sales: â‚¹{monthly_sales.max():,.2f}")
print(f"Lowest Monthly Sales: â‚¹{monthly_sales.min():,.2f}")
print(f"Average Monthly Sales: â‚¹{monthly_sales.mean():,.2f}")
```

# Customer and Product Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 1. Age Distribution

plt.figure(figsize=(10,5))
df['Age'].plot.hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

# Age summary statistics
print("Average Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Age Standard Deviation:", df['Age'].std())
```

```python
# 2. Sales by Gender

gender_sales = df.groupby('Gender')['Total Amount'].sum()
print("\nTotal Sales by Gender:\n", gender_sales)

avg_gender_spend = df.groupby('Gender')['Total Amount'].mean()
print("\nAverage Spend by Gender:\n", avg_gender_spend)

# Bar chart: Total Sales by Gender
plt.figure(figsize=(6,4))
gender_sales.plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Total Sales by Gender')
plt.ylabel('Sales (â‚¹)')
plt.tight_layout()
plt.show()
```

```python
# 3. Age Group Segmentation

df['Age Group'] = pd.cut(df['Age'], bins=[0, 20, 30, 40, 50, 60, 70], labels=['<20','21-30','31-40','41-50','51-60','61+'])

age_group_sales = df.groupby('Age Group')['Total Amount'].mean()
print("\nAverage Spend by Age Group:\n", age_group_sales)

total_age_group_sales = df.groupby('Age Group')['Total Amount'].sum()
print("\nTotal Sales by Age Group:\n", total_age_group_sales)

# Bar chart: Total Sales by Age Group
plt.figure(figsize=(8,5))
total_age_group_sales.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by Age Group')
plt.ylabel('Sales (â‚¹)')
plt.tight_layout()
plt.show()
```

```python
# 4. Product Category Analysis

popular_categories = df['Product Category'].value_counts()
print("\nProduct Category Frequency:\n", popular_categories)

category_sales = df.groupby('Product Category')['Total Amount'].sum()
print("\nTotal Sales by Category:\n", category_sales)

avg_category_spend = df.groupby('Product Category')['Total Amount'].mean()
print("\nAverage Spend by Category:\n", avg_category_spend)

# Bar chart: Total Sales by Product Category
plt.figure(figsize=(8,5))
sns.barplot(x='Product Category', y='Total Amount', data=df, estimator=sum, ci=None, palette='pastel')
plt.title('Total Sales by Product Category')
plt.tight_layout()
plt.show()
```

# Visualization - Present insights through bar charts, line plots, and heatmaps.

# 1. Bar Chart

```python
# Total Sales by Product Category

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(x='Product Category', y='Total Amount', data=df, estimator=sum, ci=None, palette='pastel')
plt.title('Total Sales by Product Category')
plt.ylabel('Total Sales (â‚¹)')
plt.tight_layout()
plt.show()
```

```python
# Total Sales by Gender

gender_sales = df.groupby('Gender')['Total Amount'].sum().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x='Gender', y='Total Amount', data=gender_sales, palette='Set2')
plt.title('Total Sales by Gender')
plt.ylabel('Total Sales (â‚¹)')
plt.tight_layout()
plt.show()
```

# 2. Line Plots â€” Show Trends Over Time

```python
# Monthly Sales Trend

df['Date'] = pd.to_datetime(df['Date'])  # Ensure date format
monthly_sales = df.resample('M', on='Date')['Total Amount'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Total Amount', data=monthly_sales, marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales (â‚¹)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
